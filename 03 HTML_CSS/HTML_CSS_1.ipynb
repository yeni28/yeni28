### HTML
태그 등을 이용해서 문서나 데이터의 구조를 명시하는 언어 => Mark up Language(ex. Markdown)

- 웹페이지를 작성(구조화)하기 위한 언어

.html 

#### HTML 요소
HTML 작성방법 : <태그이름> 포함시킬 내용</태그이름>
내용이 없는 태그들도 존재 : (닫는 태그가 없음, 또는 여는 동시에 닫는 태그가 존재)
 - br, hr, img..

HTML 요소는 중첩된다.(조건문 안에 조건문)
- 중첩을 통해서 문서를 구조화 할 수 있다.
- 여는 태그와 닫는 태그의 쌍을 잘 확인해야한다. 해당 태그의 범위가 어디까지인지 알아야 구조화가 잘된다.
- 코드를 잘못 작성하면 오류를 반환하는 게 아니라 그냥 화면이 깨지거나 오작동(원하는 대로 동작하지 않음) 디버깅이 힘들어진다.(어디가 잘못되었는지 한눈에 찾기 어렵다.)

#### HTML 태그의 속성(attribute)

```html
<a href="https://google.com"></a>
```

속성은 속성 안에 어떤 값을 넣어줄 지 표시하는 할당연산자가 있고, 보통 쌍따옴표를 통해 그 안의 값을 넣어준다.

속성작성 시 주의하는 점은 꼭 붙여서 작성!
href - 어떤 사이트로 넘어갈 지 연결해주는 속성

- 속성을 통해서 태그의 부가적인 정보를 설정할 수 있다.
- html요소는 속성을 가질 수 있다.(경로, 요소의 크기 같은 추가적인 제공 )
- 요소의 시작태그에 작성 , 보통 이름과 값이 하나의 쌍으로 존재한다.

#### 시맨틱 태그
디브 태그의 일종, 그 디브태그에 역할을 달아준 것.
동작하는 방식은 디브태그와 똑같다.


```
<header> 웹페이지의 상단에 존재하는 div태그</header>

<footer> 웹페이지의 하단에 존재하는 div태그</footer>
```

- div 태그에 역할에 따른 이름을 부여 ==> 의미론적 마크업

- 명확하게 이름을 지어주면, 여기가 웹페이지 상단에 존재하는 div태그겠군 하고 한 번에 알아볼 수 있다.
- 단순히 구역을 나누는 것에 더해 태그가 의미를 가지도록 해주는 요소
- 의미가 명확해지기 때문에 코드의 가독성을 높일 수 있다. => 유지보수가 쉬워진다.
- 검색엔진 최적화를 위해서 사용한다.
